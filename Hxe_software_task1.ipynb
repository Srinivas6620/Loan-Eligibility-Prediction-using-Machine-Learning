Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load the Iris Dataset

```python
df = pd.read_csv(r"C:\Users\teste\Downloads\IRIS.csv")
print("First 5 rows of the dataset:")
print(df.head())
```

Step 3: Understand the Dataset

```python
print("\nShape of the dataset:", df.shape)
print("\nData Types and Missing Info:")
print(df.info())
print("\nStatistical Summary:")
print(df.describe())
```

Step 4: Check for Missing Values

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

Step 5: Basic Statistical Analysis

```python
print("\nMean values grouped by species:")
print(df.groupby("species").mean())
correlation = df.drop(columns=["species"]).corr()
print("\nCorrelation Matrix:")
print(correlation)
```

Step 6: Data Visualizations

```python
sns.pairplot(df, hue='species')
plt.suptitle("Pairwise Plots of Iris Features", y=1.02)
plt.show()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length by Species")

plt.subplot(1, 2, 2)
sns.boxplot(x='species', y='petal_length', data=df)
plt.title("Petal Length by Species")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='species', y='petal_width', data=df, palette="Pastel1")
plt.title("Petal Width Distribution by Species")
plt.show()
```

```python
sns.histplot(data=df, x='sepal_length', hue='species', kde=True)
plt.title("Histogram + KDE of Sepal Length")
plt.show()
```

```python
sns.swarmplot(x='species', y='petal_length', data=df)
plt.title("Swarm Plot of Petal Length")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
species_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
colors = df['species'].map(species_map)
ax.scatter(df['sepal_length'], df['petal_length'], df['petal_width'], c=colors)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
plt.title("3D Scatter Plot")
plt.show()
```

```python
sns.countplot(x='species', data=df)
plt.title("Count of Each Species")
plt.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x='sepal_length', y='petal_length', z='petal_width',
                    color='species', title="3D Scatter with Plotly")
fig.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', y='sepal_width', hue='species', fill=True, alpha=0.5)
plt.title("2D KDE Plot of Sepal Length vs Width")
plt.show()
```

```python
g = sns.FacetGrid(df, col="species")
g.map(plt.hist, "petal_length", bins=10)
plt.suptitle("Petal Length Distribution by Species")
plt.tight_layout()
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(df, 'species', colormap='Set2')
plt.title("Parallel Coordinates Plot")
plt.show()
```

```python
mean_values = df.groupby('species').mean()
features = mean_values.columns
labels = mean_values.index

angles = np.linspace(0, 2*np.pi, len(features), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for label in labels:
    values = mean_values.loc[label].tolist()
    values += values[:1]
    ax.plot(angles, values, label=label)
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(features)
plt.title("Radar Chart of Mean Feature Values")
plt.legend()
plt.show()
```
